```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(x = "sepal_length",y="petal_length",data = iris)
```

```python
import matplotlib.pyplot as plt
plt.scatter(iris["sepal_width"],iris["petal_width"])
```

```python
sns.lineplot(x="sepal_width",y="petal_width",data = iris)
```

```python
sns.scatterplot(x = "sepal_length",y="sepal_width",data = iris,color = "yellow")
```

```python
sns.distplot(iris["sepal_length"])
```

```python
sns.boxplot(x="sepal_length",y="sepal_width",data = iris)
```

```python
sns.catplot(x="sepal_length",y = "species",data = iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot = True)
```
